# Cartopy Distortions:
## Here, we will explore how different map projections affect the way states we all know and love (like Iowa) are shown. This can be very important depending on the use case: some projections will distort distances, 
## making something like a drive on I80 seem shorter or farther than it really is. Other projections can be used in a "propaganda" map: on a mercator projection, for example, areas in higher latitudes appear higher than they really are.


```python
import matplotlib.pyplot as plt
import cartopy.feature as cfeature
import cartopy.crs as ccrs
```

```python
# Iowa. 

latN = 43.5
latS = 40.4
lonW = -96.5
lonE = -90.1

cLat = (latN + latS) / 2
cLon = (lonW + lonE) / 2
```

```python
projIA = ccrs.PlateCarree()
```

```python
fig = plt.figure(figsize= (11, 8.5))

ax = plt.axes(projection= projIA)

ax.coastlines()
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.STATES, linewidth= 0.5)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.LAND)

ax.set_extent([lonW - 1, lonE + 1, latS - 1, latN + 1], crs= projIA)
ax.set_title("Iowa with Plate Carree (yuck) projection")

plt.show()
```

```python
ax = plt.axes(projection= ccrs.LambertAzimuthalEqualArea(central_longitude= cLon, central_latitude= cLat))

ax.add_feature(cfeature.STATES, linewidth= 0.5)
ax.add_feature(cfeature.LAND)

ax.set_extent([lonW - 1, lonE + 1, latS - 1, latN + 1], crs= projIA)
ax.set_title("Iowa with Lambert Azimuthal Equal Area projection")

plt.show()
```

```python
ax = plt.axes(projection= ccrs.EckertIII(central_longitude= cLon))

ax.add_feature(cfeature.STATES, linewidth= 0.5)
ax.add_feature(cfeature.LAND)

ax.set_extent([lonW - 1, lonE + 1, latS - 1, latN + 1], crs= projIA)
ax.set_title("Iowa with EckertIII projection")

plt.show()
```

```python
ax = plt.axes(projection= ccrs.Gnomonic(central_longitude= cLon, central_latitude= cLat))

ax.add_feature(cfeature.STATES, linewidth= 0.5)
ax.add_feature(cfeature.LAND)

ax.set_extent([lonW - 1, lonE + 1, latS - 1, latN + 1], crs= projIA)
ax.set_title("Iowa with Gnomonic projection")

plt.show()
```

```python
ax = plt.axes(projection= ccrs.Stereographic(central_longitude= cLon, central_latitude= cLat))

ax.add_feature(cfeature.STATES, linewidth= 0.5)
ax.add_feature(cfeature.LAND)

ax.set_extent([lonW - 1, lonE + 1, latS - 1, latN + 1], crs= projIA)
ax.set_title("Iowa with Stereographic projection")

plt.show()
```

```python
ax = plt.axes(projection= ccrs.Mercator(central_longitude= cLon))

ax.add_feature(cfeature.STATES, linewidth= 0.5)
ax.add_feature(cfeature.LAND)

ax.set_extent([lonW - 1, lonE + 1, latS - 1, latN + 1], crs= projIA)
ax.set_title("Iowa with Mercator projection")

plt.show()
```
