```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs 
import cartopy.feature as cfeature
```

### Map with specified projection 

```python
fig = plt.figure(figsize= (5,5))

ax = plt.axes(projection = ccrs.AlbersEqualArea())
```

### Adding features using cartopy.feature

```python
ax = plt.axes(projection=ccrs.AlbersEqualArea())
ax.coastlines()
ax.add_feature(cfeature.BORDERS)
ax.stock_img()
plt.show()
```

Other map projections

```python
ax = plt.subplot(projection=ccrs.UTM(15,southern_hemisphere=False,globe=None))
```

### Regional Maps 


```python
latN = 43.5
latS = 40.4
lonW = -96.5
lonE = -90.1

cLat = (latN + latS) / 2
cLon = (lonW + lonE) / 2

projIA2 =ccrs.TransverseMercator(central_longitude=cLon)

fig, ax = plt.subplots(figsize=(8, 6), subplot_kw={'projection': projIA2})

ax.set_extent([lonW,lonE,latS,latN], crs=ccrs.PlateCarree())

ax.add_feature(cfeature.STATES, linestyles="-")

plt.title("Blank Map of Iowa")

plt.show()
```

 We can make regional maps by using set_extent() with longitude and latitude values!

```python
latN = 43.5
latS = 40.4
lonW = -96.5
lonE = -90.1

cLat = (latN + latS) / 2
cLon = (lonW + lonE) / 2

projIA = ccrs.LambertConformal(central_longitude=cLon)

fig, ax = plt.subplots(figsize=(8, 6), subplot_kw={'projection': projIA})


ax.set_extent([lonW,lonE,latS,latN], crs=ccrs.PlateCarree())

# ax.add_feature(cfeature.COASTLINE)
# ax.add_feature(cfeature.BORDERS, linestyle=":")
ax.coastlines(resolution='10m')
ax.add_feature(cfeature.STATES, linestyles="-")
ax.add_feature(cfeature.BORDERS, linestyles=':')

ax.stock_img()
plt.title("Corn State") # corn

plt.show()
```

### Cartopy can also be used to plot maps side by side using .subplot()


```python
fig = plt.figure(figsize=(10,5))

ax1 = plt.subplot(1,2,1, projection=ccrs.InterruptedGoodeHomolosine())
ax2 = plt.subplot(1,2,2, projection=ccrs.Mollweide())

# now add features!

#coastlines
ax1.add_feature(cfeature.COASTLINE)
ax2.add_feature(cfeature.COASTLINE)

ax1.add_feature(cfeature.OCEAN)
ax2.add_feature(cfeature.OCEAN)

ax1.add_feature(cfeature.LAND)
ax2.add_feature(cfeature.LAND)

plt.show()
```
