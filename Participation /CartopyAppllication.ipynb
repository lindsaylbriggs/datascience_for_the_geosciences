# Cartopy Applications 

Make a graph of storm reports for the state of Iowa. Make up data as you please. 

```python
import matplotlib.pyplot as plt 
import cartopy.crs as ccrs 
import cartopy.feature as cfeature 

latN = 43.5  # Northern boundary
latS = 40.4  # Southern boundary
lonW = -96.5  # Western boundary
lonE = -90.1  # Eastern boundary

```

### Plot data onto a map

```python
fig = plt.figure(figsize= (11, 8.5))

ax = plt.axes(projection= ccrs.AlbersEqualArea(central_longitude= ((lonW + lonE) / 2)))

ax.set_extent([lonW, lonE, latS, latN], crs= ccrs.PlateCarree())

ax.add_feature(cfeature.STATES)


ax.scatter(-93.625, 41.5868, color= 'firebrick', transform= ccrs.PlateCarree())
ax.scatter(-91.6656, 41.9779, color= 'firebrick', transform= ccrs.PlateCarree())
ax.scatter(-90.5776, 41.5236, color= 'firebrick', transform= ccrs.PlateCarree())
ax.scatter(-93.6199, 42.0347, color= 'firebrick', transform= ccrs.PlateCarree())
ax.scatter(-96.6199, 42.5006, color= 'firebrick', transform= ccrs.PlateCarree())
ax.scatter(-92.225, 42.5598, color= 'navy', transform= ccrs.PlateCarree())
ax.scatter(-94.962, 41.596, color= 'navy', transform= ccrs.PlateCarree())



plt.show()
```

### What about a legend?

```python
import matplotlib.patches as mpatches

fig = plt.figure(figsize= (11, 8.5))

ax = plt.axes(projection= ccrs.AlbersEqualArea(central_longitude= ((lonW + lonE) / 2)))

ax.set_extent([lonW, lonE, latS, latN], crs= ccrs.PlateCarree())

ax.add_feature(cfeature.STATES)


ax.scatter(-93.625, 41.5868, color= 'firebrick', label= 'Tornado', transform= ccrs.PlateCarree())
ax.scatter(-91.6656, 41.9779, color= 'firebrick', label= 'Tornado', transform= ccrs.PlateCarree())
ax.scatter(-90.5776, 41.5236, color= 'firebrick', label= 'Tornado', transform= ccrs.PlateCarree())
ax.scatter(-93.6199, 42.0347, color= 'firebrick', label= 'Tornado', transform= ccrs.PlateCarree())
ax.scatter(-96.6199, 42.5006, color= 'firebrick', label= 'Tornado', transform= ccrs.PlateCarree())
ax.scatter(-92.225, 42.5598, color= 'navy', label= 'Hail',transform= ccrs.PlateCarree())
ax.scatter(-94.962, 41.596, color= 'navy', label= 'Hail', transform= ccrs.PlateCarree())

ax.legend(loc= 'upper right', title= 'Storm Reports')

ax.set_title('Storm Reports in Iowa on April 1, 10 BCE')

plt.show()
```

### Okay, so now we have a legend. But I don't like how the events repeat. Can you fix that?

Yes! It can be fixed.
